# Slicing and selecting specific data from your DataFrame
**loc()** *and* **iloc()** methods help us collect rows , columns and specific data values from our DataFrame

```python
result = car_sales.loc[9, 'Colour']
print(result) #the color of the vehichle with the index number 9 
```

```python
result = car_sales.loc[:, ['Make', 'Price']]
print(result ) # select all rows from the two columns named 'Make ' ,' Price'
```

```python
r = car_sales.loc[ [0, 1, 5], :]
print(r)# select all columns from the rows with the indexes   0 , 1 , 5 
```

```python
r1 = car_sales.loc[ 5:8 , :] 
#We select all the columns , from the rows numbered 5 to 8 
r1
```
